# Contagem regressiva maluca

Estamos habituados com contagens regressivas. Fazemos contagem regressiva para a chegada de um ano novo, uma data especial, o começo de um filme, dentre outras atividades. Existem diversos aplicativos no mercado com este fim, chamados temporizadores. Uma característica comum a muitos deles é que, a medida que o prazo final se aproxima, a atualização do tempo fica mais frequente. Por exemplo, primeiro começam reportando o tempo em dias, depois em horas, em minutos, e finalmente em segundos. Essa característica deixa o aplicativo mais útil pois nos permite mensurar melhor o tempo que falta até o prazo final.

Um bilionário chamado John Williams precisa de um temporizador para o seu iphone de ouro. O Sr. Williams quer algo exclusivo, porém mais do que um aplicativo desenvolvido exclusivamente para ele, ele quer algo diferente de tudo que existe no mercado (exclusividade é seu nome!). Após consulta com alguns especialistas da UFABC, notou que este é um mercado saturado e que todos os temporizadores que poderiam ser desenvolvidos já o foram, percebeu que a única forma de obter exclusividade seria se o seu temporizador fosse inútil.

Assim, Sr. Williams deseja um temporizador de segundos que passe a atualizar o tempo mais raramente. Ele quer que o intervalo entre cada atualização fique dois segundos maior após cada uma delas. Por exemplo, se ele iniciar o temporizador com 50 segundos, então receberá atualizações dizendo que faltam 50, 48, 44, 38, 30, 20 e 8 segundos (note que os intervalos entre as notificações foram 2, 4, 6, 8, 10 e 12 segundos).

Sabendo do funcionamento de tal aplicativo, desenvolva um programa que exiba em quais segundos o Sr. Williams receberá atualizações, dado que o programa tenha sido inicializado com um tempo igual a N.

- Entrada:
O seu programa deve receber um número inteiro positivo N, que é o tempo inicial do temporizador.

- Saída:
Seu programa deve escrever a saída conforme os exemplos abaixo.

- Exemplos:
 - 10 -> Faltam 10 segundos Faltam 8 segundos Faltam 4 segundos Acabou

```python
tempo_restante = float(input())
contador = 0

print('Faltam %.f segundos'%tempo_restante)
while tempo_restante > 0:
  contador = contador + 2
  tempo_restante = tempo_restante - contador
  if tempo_restante > 0:
    print('Faltam %.f segundos'%tempo_restante)

print('Acabou')
```

Crie um programa que permita imprimir uma representação de um tabuleiro de xadrez.

Entrada:

Saída:
- O programa deve desenhar o tabuleiro com dois caracteres que representam as divisões em cores diferentes conforme exemplos apresentados a seguir.

Exemplos:
- 3 -> oxo xox oxo
- 4 -> oxox xoxo oxox xoxo

```python
def juntando_strings(lista_de_strings):
  string_final = ''
  for string in lista_de_strings:
    string_final = string_final + string  

  return string_final
```

```python
tamanho_do_tabuleiro = float(input())

tabuleiro = []

count = 3
for i in range(int(tamanho_do_tabuleiro)):
  linha = []  
  for i in range(int(tamanho_do_tabuleiro)):
    if count % 2 == 0: # par
      linha.append('x')
    else:
      linha.append('o')
    
    count = count + 1
  
  if tamanho_do_tabuleiro % 2 == 0:
    count = count + 1


  tabuleiro.append(juntando_strings(linha))

# print final
for linha in tabuleiro:
  print(linha)
```

# Falha do Motor

Ao observar a curva de velocidade de um motor, um engenheiro percebeu que sempre ocorria uma queda de velocidade quando medições eram feitas em intervalos de 10 ms. Após realizar alguns testes, ele observou que tais quedas não ocorriam necessariamente no mesmo momento. Intrigado pela falta de padrão, agora ele quer a sua ajuda para saber: dado um caso de teste, qual a primeira medição em que ocorreu uma queda de velocidade?

Entrada:
- A primeira linha contém um número inteiro N (1 < N ≤ 100) representando a quantidade de medições de velocidade do motor em um determinado teste. Cada uma das próximas N linhas consiste de um único inteiro M (0 ≤ M ≤ 10000) representando o número de RPM (rotações por minuto) daquela medida.

Saída:
- A saída é o índice da medição em que ocorreu a primeira queda de velocidade. Caso não aconteça nenhuma queda, o seu programa deve imprimir o número 0.

Exemplos:
- 3 1 4 2 3 -> 3
- 5 100 199 199 198 0 -> 4


```python
quantidade_de_medicoes = float(input())

lista_de_medicoes = []

for i in range(int(quantidade_de_medicoes)):
  medicao = float(input())
  lista_de_medicoes.append(medicao)


posicao_das_quedas = []
for i, medicao in enumerate(lista_de_medicoes):
  if i == 0: pass
  else:
    if medicao < lista_de_medicoes[i - 1]:
      posicao_das_quedas.append(i)


if posicao_das_quedas == []:
  print(0)
else:
  print(posicao_das_quedas[0] + 1)
```

# Pula Sapo

Pula sapo trata-se de um jogo onde em cada fase o jogador deve conduzir um pequeno anfíbio através de uma sequência de canos de diferentes alturas até chegar a salvo no cano mais à direita. Entretanto, o sapo só consegue sobreviver se a diferença de altura entre canos consecutivos for de, no máximo, a altura do pulo do sapo.

Caso a altura do cano seguinte for muito alta, o sapo bate no cano e cai.

Se a altura do cano seguinte for muito baixa, o sapo não aguenta a queda.

O sapo sempre começa em cima do cano mais à esquerda.

image.png

A distância entre os canos é irrelevante, ou seja, o sapo sempre consegue alcançar o próximo cano com um pulo.

Sua tarefa é escrever um programa de computador que, dadas as alturas dos canos e a altura do pulo do sapo, mostra se a fase do jogo pode ser vencida ou não.

Entrada:
- O programa recebe vários números inteiros: Os primeiros dois inteiros positivos P e N correspondem ao pulo do sapo e o número de canos. Os demais N inteiros positivos indicam as alturas dos canos ordenados da esquerda para a direita.

Saída:
- O programa deve imprimir uma mensagem: Se o sapo pode chegar no cano mais à direita, escreva YOU WIN. Se o sapo não consegue, escreva GAME OVER. Com todas as letras em maíúsculo.

Exemplos:
- 1 2 2 2 -> YOU WIN
- 1 2 1 3 -> GAME OVER

```python
alcance_do_pulo = float(input())
numero_de_canos = float(input())


lista_de_alturas_de_canos = []
for i in range(int(numero_de_canos)):
  lista_de_alturas_de_canos.append(float(input()))


lista_de_pulos = []
for index, altura_atual in enumerate(lista_de_alturas_de_canos[:-1]):
  delta = (altura_atual - lista_de_alturas_de_canos[index + 1]) # comparando cano atural com o seguinte...
  print(delta)
  # if delta < 0: #  se for menor que 0 (negativo), ver se é maior ou menor que alcance do pulo
  if abs(delta) <= alcance_do_pulo:
    lista_de_pulos.append('YOU WIN')
  else:
    lista_de_pulos.append('YOU LOSE')
  # else: lista_de_pulos.append('YOU WIN')


if 'YOU LOSE' in lista_de_pulos: print('GAME OVER')
else: print('YOU WIN')
```

# Street Fighter

O jogo Street Fighter foi um dos primeiros jogos eletrônicos do gênero conhecido como jogos de luta. Neste jogo, um mestre de artes marciais, chamado Ryu, enfrenta outros lutadores em um torneio internacional de artes marciais. Cada combate entre Ryu e um oponente se dá em uma série de rounds. No início de cada round, cada jogador está com 100 pontos de vida. O objetivo é atacar o oponente com diferentes golpes, sendo que cada golpe aplicado subtrai uma certa quantidade de pontos de vida do outro combatente. Perde o round aquele jogador cujos pontos de vida chegar primeiro à zero. Vence a luta quem ganhar o maior número de rounds.

image.png

Crie um programa que simule uma luta entre Ryu e Ken e determine quem ganhou a luta.

Entrada
- A entrada consiste de uma sequência de inteiros, um em cada linha, representando os valores dos golpes aplicados (valores positivos) e recebidos (valores negativos) por Ryu. A luta termina quando o inteiro lido for 0. Um round só termina quando um dos jogadores fica com 0 ou menos pontos de vida. Você pode assumir que em cada round pelo menos um golpe será aplicado.

Saída
- A saída deverá conter apenas uma linha, contendo somente “Ryu venceu”, “Ken venceu” ou “Empate”, de acordo com o resultado geral da luta.

Exemplo comentado
- Suponha que a sequência de entrada seja composta pelos números: 40, 5, -32, 9, -45, 48, -27, -64, 75, 6, 4, -7, 3, -39, 0.
- Então temos dois rounds. O primeiro round tem 6 golpes e Ryu ganha, pois 40 + 5 + 9 + 48 = 102 ≥ 100. O segundo round tem 8 golpes e Ken ganha, pois 27 + 64 + 7 + 39 = 137 ≥ 100. Logo, o resultado da luta é “Empate”.

```python
def fight(lista_de_golpes):
  vida_ryu, vida_ken = 100, 100
  ryu_wins, ken_wins = 0, 0

  for golpe in lista_de_golpes:
    if golpe > 0: # golpe desferido por ryu
      vida_ken = vida_ken - abs(golpe)
    else: # golpe desferido por ken
      vida_ryu = vida_ryu - abs(golpe)

    # vida de alguém chegou a 0... impondo vitória e resetando vida dos players
    if vida_ryu <= 0:
      ken_wins = ken_wins + 1
      vida_ryu, vida_ken = 100, 100
    elif vida_ken <= 0:
      ryu_wins = ryu_wins + 1
      vida_ryu, vida_ken = 100, 100
    
  return ryu_wins, ken_wins
```

```python
lista_de_golpes = [999]
while lista_de_golpes[-1] != 0:
  lista_de_golpes.append(float(input()))
lista_de_golpes = lista_de_golpes[1:]


ryu_wins, ken_wins = fight(lista_de_golpes)

if ryu_wins == ken_wins: print('Empate')
elif ryu_wins > ken_wins: print('Ryu venceu')
else: print('Ken venceu')
```

# Sexta-feira 13

A Sexta-feira no dia 13 de qualquer mês é considerada popularmente como um dia de azar. O número 13 é considerado de má sorte. Na numerologia o número 12 é considerado de algo completo, como por exemplo: 12 meses no ano, 12 tribos de Israel, 12 apóstolos de Jesus ou 12 constelações do Zodíaco. Já 13 é considerado um número irregular, sinal de infortúnio. Assim, tem-se pela tradição, que a sexta-feira dia 13 é o mais azarado dos dias.

Triscaidecafobia é um medo irracional ao número 13. O medo específico da sexta-feira 13 é chamado de parascavedecatriafobia ou frigatriscaidecafobia.

Dado um número inteiro que representa um ano entre 1583 e 3000, faça um programa que determine a quantidade de sextas-feiras 13 que o ano dado possui. (curiosidade: 1583 é o 1º ano completo do calendário Gregoriano vigente, que foi introduzido em 24/02/1582)

Dica 1: Contabilize o número de dias decorridos desde 01/01/1583 até o dia 01/01 do ano recebido como entrada, sabendo que 01/01/1583 é um Sábado. Isso permite determinar facilmente o dia da semana do dia 01/01 de um ano dado.

Dica 2: Ao contabilizar o número de dias decorridos desde 01/01/1583, cuidado com os anos bissextos! Um ano é bissexto se a seguinte expressão for verdadeira:

(ano módulo 4 = 0 e ( ano módulo 400 = 0 ou ano módulo 100 ≠ 0 ))

Dica 3: Em um dado mês teremos sexta-feira 13, se o dia 1º inicia em um Domingo. Após determinar o dia da semana do dia 01/01 do ano recebido como entrada (ver dicas anteriores), isso permite determinar facilmente quais meses do ano dado caem em um domingo (lembrando de checar se o ano dado é bissexto, conforme expressão dada na Dica 2).

Entrada:
- O programa recebe um número inteiro que representa um ano entre 1583 e 3000.

Saída:
- O programa deve imprimir um número inteiro representando a quantidade de sextas-feiras 13 que o ano possui.

Exemplos:

- 2019 -> 2
- 2002 -> 2
- 2000 -> 1
- 1998 - 3

```python
def checando_bissexto(ano):
  if ano % 4 == 0 and ano % 100 != 0:
    return True
  elif ano % 400 == 0:
    return True
  else:
    return False


def contagem_de_dias_decorridos(primeiro_ano, ultimo_ano):
  count_dias = 0
  for ano in range(primeiro_ano, ultimo_ano, 1):
    if checando_bissexto(ano) == True: 
      dias_no_ano = 366
    else:
      dias_no_ano = 365

    count_dias += dias_no_ano
  
  return count_dias


def dia_da_semana_do_primeiro_dia_do_ano(ano):# checando dia da semana
  dia_da_semana = 7 # 01/01/1583 era um Sábado (dia 7 da semana)
  for dia in range(contagem_de_dias_decorridos(1583, ano)):
    if dia_da_semana == 7:
      dia_da_semana = 1
    else:
      dia_da_semana += 1

  return dia_da_semana


def contando_sexta_feira_13(ano_escolhido):
  dias_no_mes = {1: 31, 2: 28, 3: 31, 4: 30, 5: 31, 6: 30, 7: 31, 8: 31, 9: 30, 10: 31, 11: 30, 12: 31} # 1 -> janeiro, 2 -> fevereiro... 12 -> dezembro

  if checando_bissexto(ano_escolhido) == True:  
    dias_no_ano = 366
    dias_no_mes[2] = 29
  else:  dias_no_ano = 365

  dia = 1
  dia_da_semana = dia_da_semana_do_primeiro_dia_do_ano(ano_escolhido) # 1 -> domingo, 2 -> segunda, 3 -> terça, 4 -> quarta, 5 -> quinta, 6 -> sexta, 7 -> sabado
  mes = 1
  count_sexta_feira_treze = 0
  for i in range(dias_no_ano):
    if dia <= dias_no_mes[mes]:
      if dia == 13 and dia_da_semana == 6: # sexta-feira treze
        count_sexta_feira_treze += 1
    else:
      dia = 1
      mes += 1

    dia += 1
    if dia_da_semana == 7: dia_da_semana = 1
    else: dia_da_semana += 1
  
  return count_sexta_feira_treze
```

```python
ano_escolhido = int(input())
print(contando_sexta_feira_13(ano_escolhido))
```

# Miojo

João é um fanático por miojos; ele os adora, e, como era de se esperar, ele levou vários pacotes quando foi acampar com seus colegas. Como João só gosta de miojos feitos com o tempo exato, ele se desesperou ao perceber que havia esquecido seu relógio em casa.

Por sorte, ele conseguiu, no caminho, comprar duas ampulhetas de durações diferentes. Por exemplo, se o miojo precisa de 3 minutos para ficar pronto, e João tiver uma ampulheta de 5 minutos e outra de 7, uma possível forma de cozinhar o miojo é:

1. João começa virando as duas ampulhetas ao mesmo tempo;

2. Quando a areia da ampulheta de 5 minutos se esgotar, João torna a virá-la;

3. João começa a ferver o miojo quando a areia da ampulheta de 7 minutos acabar;

4. João tira o miojo do fogo quando a ampulheta de 5 minutos acabar novamente.

Dessa forma, o miojo ficará exatamente 3 minutos no fogo (do minuto 7 ao minuto 10). Assim, apesar do miojo levar apenas três minutos para ser cozido, ele precisa de 10 minutos para ficar pronto.

Faça um programa que, dado o tempo de preparo do miojo, e os tempos das duas ampulhetas (ambos maiores que o tempo do miojo), determina o tempo mínimo necessário para o miojo ficar pronto.

Você pode supor que sempre é possível cozinhar o miojo no tempo correto.

Entrada:
- A entrada contém três inteiros T, A e B, representando o tempo necessário para o preparo do miojo, o tempo da primeira ampulheta e o tempo da segunda ampulheta, respectivamente.

Observe que: 0 ≤ T ≤ 10000 e T < A,B ≤ 40000.

Saída:
- Seu programa deve imprimir o tempo mínimo para o preparo do miojo.

Exemplos:
- 3 5 7 -> 10
- 14 15 22 -> 44

```python
def tempo_para_preparar_miojo_com_ampulhetas(miojo_tempo, ampulheta_a, ampulheta_b):
  count_time = 0
  temp_ampulheta_a = ampulheta_a
  temp_ampulheta_b = ampulheta_b

  delta = []
  while delta != 0:
    if temp_ampulheta_a > temp_ampulheta_b:
      delta = temp_ampulheta_a - temp_ampulheta_b
      count_time += temp_ampulheta_b
      temp_ampulheta_a = delta
      temp_ampulheta_b = ampulheta_b
    else:
      delta = temp_ampulheta_b - temp_ampulheta_a
      count_time += temp_ampulheta_a
      temp_ampulheta_a = ampulheta_a
      temp_ampulheta_b = delta

    if delta == miojo_tempo:
      count_time += miojo_tempo
      return count_time
```

```python
miojo_tempo = float(input())
ampulheta_a = float(input())
ampulheta_b = float(input())

print(int(tempo_para_preparar_miojo_com_ampulhetas(miojo_tempo, ampulheta_a, ampulheta_b)))
```
