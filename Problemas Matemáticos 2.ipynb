# Intervalo

Faça um programa que leia um número qualquer e apresente uma mensagem dizendo em qual dos seguintes intervalos ([0,25], (25,50], (50,75], (75,100]) este valor se encontra. Se o valor não estiver em nenhum destes intervalos, deverá ser impressa a mensagem “Fora de intervalo”.

O símbolo ( representa "maior que". Por exemplo:

- [0,25] indica valores entre 0 e 25, inclusive eles.
- (25,50] indica valores maiores que 25 (Ex: 25.00001) até o valor 50.

Entrada:
- A entrada contém um número (inteiro ou real).

Saída:
- Seu programa deve imprimir na tela uma mensagem conforme exemplos abaixo.

Exemplos:
- 25.01 -> Intervalo (25, 50]
- 25 -> Intervalo [0, 25]
- 100 -> Intervalo (75, 100]
- -25.02 > Fora de intervalo

```python
x = float(input('Digite o número: '))

if x >= 0 and x <= 25:
  print('Intervalo [0, 25]')
elif x > 25 and x <= 50:
  print('Intervalo (25, 50]')
elif x > 50 and x <= 75:
  print('Intervalo (50, 75]')
elif x > 75 and x <= 100:
  print('Intervalo (75, 100]')
else:
  print('Fora de intervalo')
```

# Triângulos

Leia 3 valores reais A, B e C tais que  A >= B >= C. A seguir, determine o tipo de triângulo que estes três lados formam, com base nos seguintes casos:
- Se A >= B+C, apresente a mensagem “NAO FORMA TRIANGULO”
- Se A² = B² + C², apresente a mensagem “TRIANGULO RETANGULO”
- Se A² > B² + C², apresente a mensagem “TRIANGULO OBTUSANGULO”
- Se A² < B² + C², apresente a mensagem “TRIANGULO ACUTANGULO”

Se os três lados forem iguais, apresente a mensagem “TRIANGULO EQUILATERO”

Se apenas dois dos lados forem iguais, apresente a mensagem “TRIANGULO ISOSCELES”

Entrada:
- A entrada contém três valores reais positivos A, B e C, tais que  A >= B >= C, correspondentes aos três lados de um triângulo.

Saída:
- Imprima todas as classificações do triângulo especificado na entrada. Note que um triângulo pode ter mais de uma classificação.

Exemplos:
- 7.0 7.0 5.0 -> TRIANGULO ACUTANGULO TRIANGULO ISOSCELES
- 10.0 6.0 6.0 -> TRIANGULO OBTUSANGULO TRIANGULO ISOSCELES
- 6.0 6.0 6.0 -> TRIANGULO ACUTANGULO TRIANGULO EQUILATERO
- 7.0 5.0 2.0 -> NAO FORMA TRIANGULO
- 10.0 8.0 6.0 -> TRIANGULO RETANGULO

```python
A = float(input('Qual o tamanho do lado A? '))
B = float(input('Qual o tamanho do lado B? '))
C = float(input('Qual o tamanho do lado C? '))

if A >= B + C:
  print('NAO FORMA TRIANGULO')
elif A**2 == B**2 + C**2:
  print('TRIANGULO RETANGULO')
elif A**2 > B**2 + C**2:
  print('TRIANGULO OBTUSANGULO')
elif A**2 < B**2 + C**2:
  print('TRIANGULO ACUTANGULO')

if A == B:
  if B == C:
    print('TRIANGULO EQUILATERO')
  else:
    print('TRIANGULO ISOSCELES')
elif B == C:
  if A == B:
    print('TRIANGULO EQUILATERO')
  else:
    print('TRIANGULO ISOSCELES')
```

Entrada
- A entrada contém 6 números reais representando as coordenadas dos pontos: a1, a2, b1, b2, c1, c2.

Saída
- Uma das três strings: a, b, empate.

Exemplos:
- 0, 0, 1, 1, 2, 2 -> b
- 1, 1, 0, 0, 2, 2 -> a
- 3, 3, 1, 1, 2, 2 -> empate
- 2.5, 7.5, -3.2, 6.4, -0.1, -8 -> b
- 51, 16, -1, -25, 15, 13 ->

image.png

```python
import math

a1 = float(input('Ponto a1: '))
a2 = float(input('Ponto a2: '))
b1 = float(input('Ponto b1: '))
b2 = float(input('Ponto b2: '))
c1 = float(input('Ponto c1: '))
c2 = float(input('Ponto c2: '))

distancia_entre_a_e_c = math.sqrt((c1 - a1)**2 + (c2 - a2)**2)
distancia_entre_b_e_c = math.sqrt((c1 - b1)**2 + (c2 - b2)**2)

if distancia_entre_a_e_c < distancia_entre_b_e_c:
  print('a')
elif distancia_entre_a_e_c > distancia_entre_b_e_c:
  print('b')
else:
  print('empate')
```

# Balanceamento de Carga na Aliança Rebelde



O transporte interestelar de cargas apresenta uma série de medidas de segurança, e uma dessas medidas é a distribuição de pesos no interior da espaçonave. Essa medida estabelece que deve haver um balanceamento na carga que será transportada, ou seja, o peso na parte traseira da aeronave deve ser compatível com o peso na parte da frente da aeronave. Quando o balanceamento é bem realizado, a espaçonave voa mais rápido e consome menos combustível.

Nas bases de abastecimento da aliança rebelde, cada espaçonave é carregada com 4 containers, que estão carregados com munições e suprimentos. Cada container apresenta um certo peso total em toneladas, representado por um número inteiro positivo.

Como engenheiro de software júnior da aliança rebelde (porque não é qualquer um que pode ser um padawan), você deve criar um programa que, dado o peso de 4 containers, determina se é possível estabelecer um balanceamento da carga perfeito, ou seja, se é possível dividir a carga na parte traseira e dianteira da espaçonave de tal forma que o peso na parte traseira seja igual ao peso na parte dianteira.

image.png

```python
def balancear_metade(carga1, carga2, carga3, carga4):
  base1 = carga1 + carga2
  base2 = carga3 + carga4

  if base1 == base2:
    return 'balanceavel'
  else:
    return 'nao balanceavel'

def balancear_desigual(carga1, carga2, carga3, carga4):
  base1 = carga1 + carga2 + carga3
  base2 = carga4

  if base1 == base2:
    return 'balanceavel'
  else:
    return 'nao balanceavel'
```

```python
carga1 = 13
carga2 = 33
carga3 = 19
carga4 = 65

possibilidades = []

# metade-metade
possibilidades.append(balancear_metade(carga1, carga2, carga3, carga4))
possibilidades.append(balancear_metade(carga1, carga3, carga2, carga4))
possibilidades.append(balancear_metade(carga1, carga4, carga2, carga3))

# três-um
possibilidades.append(balancear_desigual(carga1, carga2, carga3, carga4))
possibilidades.append(balancear_desigual(carga2, carga3, carga4, carga1))
possibilidades.append(balancear_desigual(carga1, carga3, carga4, carga2))
possibilidades.append(balancear_desigual(carga1, carga2, carga4, carga3))

if 'balanceavel' in possibilidades:
  print('balanceavel')
else:
  print('nao balanceavel')
```

# Tempo de Jogo

Leia a hora inicial, minuto inicial, hora final e minuto final de um jogo. A seguir calcule a duração do jogo, sabendo que o mesmo pode começar em um dia e terminar em outro, tendo uma duração mínima de 1 minuto e duração máxima de 24 horas.

Entrada
- A entrada contém quatro valores inteiros representando a hora inicial, o minuto inicial, a hora final e o minuto final.

Saída
- Apresente a duração do jogo em horas e minutos conforme exemplos abaixo.


Exemplos:
- 7 10 9 20 -> 2h10m
- 15 30 17 10 -> 1h40m
- 20 50 0 24 -> 3h34m
- 23 59 0 0 -> 0h1m
- 13 13 13 13 -> 24h0m

```python
hora_inicial = float(input())
minuto_inicial = float(input())
hora_final = float(input())
minuto_final = float(input())

inicial_em_minutos = hora_inicial * 60 + minuto_inicial
final_em_minutos = hora_final * 60 + minuto_final


if inicial_em_minutos == final_em_minutos:
  print('24h0m')
elif inicial_em_minutos > final_em_minutos:
  duracao_minutos = (24 * 60) - (inicial_em_minutos - final_em_minutos)
  hora = duracao_minutos // 60
  minuto = duracao_minutos % 60
  print('%.fh'%hora + '%.fm'%minuto)
else:
  duracao_minutos = final_em_minutos - inicial_em_minutos
  hora = duracao_minutos // 60
  minuto = duracao_minutos % 60
  print('%.fh'%hora + '%.fm'%minuto)
```
